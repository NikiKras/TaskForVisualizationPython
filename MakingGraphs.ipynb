```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (16,6)
df = pd.read_csv('C:/Users/1/Documents/insurance.csv',encoding='1251')
sns.pairplot(data=df,hue='smoker')
plt.savefig('Our_graph1.png') # we save the first graph
plt.show()

sns.scatterplot(data=df, x='bmi', y='charges', hue='smoker', style='sex', palette=['red','green'],size='charges')
plt.title("Зависимость размера страхового сбора от индекса массы тела", size=15, c='purple', y = 1.03)
plt.legend(fontsize = '7', loc="right", bbox_to_anchor=(1.15, 0.63), prop={'size': 15}, markerscale=2.5, labelspacing = 2.3)
plt.xlabel("Индекс массы тела", size=15)
plt.ylabel("Страховой взнос")
plt.savefig('Our_graph2.png')
plt.show()

sns.lmplot(data=df, x="bmi", y="charges", hue="smoker", height=10, legend = False)
plt.xlabel("Индекс массы тела", size = 25)
plt.ylabel("Страховой сбор", size = 25)
plt.title("Зависимость размера страхового сбора от индекса массы тела", size = 27, y = 1.05, loc = 'left')
plt.legend(fontsize = '7', loc="right", bbox_to_anchor=(1.24, 0.929), prop={'size': 20},
           title = "Курильщик", title_fontsize = '20')
plt.savefig('Our_graph3.png')
plt.show()

g = sns.lmplot( data=df, x="bmi", y="charges", hue="smoker", height=8, col ='region', row = "sex", aspect=.8, legend = False)
plt.legend(fontsize = '20')
plt.savefig('Our_graph4.png') 
plt.show()
```
